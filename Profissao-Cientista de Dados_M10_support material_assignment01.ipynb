````python
sns.pairplot(df_num, hue='qtd_filhos', palette='pastel')
plt.tight_layout()
````

````python
sns.heatmap(df.corr(), annot=True, cmap='RdYlGn', square=False, fmt=".1f", linewidth=.5)
````

2) Avalie um gráfico de dispersão (*scatterplot*) específico para as duas variáveis mais correlacionadas com *renda*.

````python
# Renda: posse_de_veiculo e tempo_emprego

features = ['posse_de_veiculo', 'tempo_emprego']
for i in range(len(features)):
    plt.figure(figsize=(12,6))
    plt.subplot(2,2, i+1 )
    sns.scatterplot(data=df, x=df['renda'], y=df[features[i]])
    plt.xlabel('renda')
    plt.tight_layout()
````

3) Na su aopinião, há outliers na variável renda?

````python
sns.scatterplot(data=df, x=df['renda'], y=df['posse_de_veiculo'])
````

4) Calcule o logaritmo da variável renda e repita a sequência de análise

````python
df['renda'] = np.log(df['renda'])


# Renda: posse_de_veiculo e tempo_emprego

features = ['posse_de_veiculo', 'tempo_emprego']
for i in range(len(features)):
    plt.figure(figsize=(12,6))
    plt.subplot(2,2, i+1 )
    sns.scatterplot(data=df, x=df['renda'], y=df[features[i]])
    plt.xlabel('renda')
    plt.tight_layout()
````

A aplicação do log aprimorou a análise, afinal, tornou a distribuição do tempo de casa em renda melhor para análise.
